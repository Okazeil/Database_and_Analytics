```python
import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/Okazeil/Database_and_Analytics/main/Energy_dataset.csv')

# Display the first few rows of the dataframe to understand its structure
print(df.head())
```

```python
import sqlite3

# Connect to a SQLite database (in-memory)
conn = sqlite3.connect(':memory:')

# Create a cursor object
cur = conn.cursor()

# Use pandas to create a table from the DataFrame
df.to_sql('energy', conn, if_exists='replace', index=False)
```

SELECT Records from the Table

```python
# Execute a SELECT statement
cur.execute("SELECT * FROM energy LIMIT 5")

# Fetch and display the results
rows = cur.fetchall()
for row in rows:
    print(row)
```
